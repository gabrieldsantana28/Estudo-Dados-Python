1. Valores: min=101, mean=301.13, max=500
   - Faz sentido para numeração de quartos de hospital?
   - Que tipo de padrão você observa nos números?

2. Std = 115.24
   - Razão std/mean = 115.24 ÷ 301.13 ≈ ?
   - Comparado com Age (38%) e Billing Amount (56%), como está?

3. Distribuição geral:
   - Olhando min=101, max=500: quantos quartos diferentes existem?
   - A média estar em 301 faz sentido?

-------------------------------------------------------------------

Respostas:

1. Faz sentido considerarmos que o primeiro quarto é o 101 e último é 500, com isso podemos ter a percepção que a média em 300 fica no meio do valores de ambos os quartos. Isto significa que temos registros consistentes, não só por este fato que exemplifiquei, mas também pela mediana ser 302.

2. Nosso std é moderadamente alto, 38,3%, comparando com 'Age' e 'Billing Amount', podemos dizer que mantém um padrão dos stds dos dados até o momento.

3. Para ter um dado preciso de quantos quartos diferentes existem, seria mais fácil realizar a contagem distinta para cada um. Não posso fornecer algo preciso, analisando pelo mínimo e máximo descrito. Porém a média estar em 301 faz sentido, baseado nos dados analisados até o momento.

```python
# Verificar quartos únicos
print(f"Total de quartos possíveis (min-max): {df['Room Number'].max() - df['Room Number'].min() + 1}")
print(f"Total de quartos únicos usados: {df['Room Number'].nunique()}")
print(f"Porcentagem de ocupação: {df['Room Number'].nunique() / (df['Room Number'].max() - df['Room Number'].min() + 1) * 100:.1f}%")
```

Com este código, podemos confirmar a quantidade de quartos.
